```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data=sns.load_dataset('titanic')
```

```python
data.head()
```

```python
# Problem 1 : Plot a Heatmap on the correlation between multiple columns
```

```python
a=data.corr()
sns.heatmap(a,cbar=False,linewidth=1.5,annot=True,fmt='.1g',cmap='summer')
```

```python
# Problem 2 : Plot the distribution of age column using a Boxplot on the basis of sex column
```

```python
sns.catplot(x='sex',y='age',kind='box',data=data)
```

```python
# Problem 3 : Plot the distribution of age using a ViolinPlot on the basis of both Pclass and Sex column
```

```python
#on thr basis of pclass
sns.catplot(x='pclass',y='age',kind='violin',data=data)
```

```python
#on the basis of sex column
sns.catplot(x='sex',y='age',kind='violin',split=False,data=data)
```

```python
# Problem 4 : Plot the count of Survivors and Non-Survivors
```

```python
sns.catplot(x='survived',kind='count',hue='sex',data=data)
```

```python
# Problem 5: Plot the count of survivors as well as non-survivors on the basis of Pclass
```

```python
sns.catplot(x='survived',kind='count',data=data,hue='pclass')
```

```python
# Problem 6: Plot a scatterplot between age and fare column on the basis of both Pclass and Sex column
```

```python
#on the basis of pclass
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
sns.relplot(x='age',y='fare',hue='pclass',style='sex',data=data,)
```
